```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text_data(df, text_column):
    """
    Preprocess text data in the specified column of the DataFrame.

    Parameters:
    df (pd.DataFrame): DataFrame containing the text data.
    text_column (str): The column name containing the raw text messages.

    Returns:
    pd.Series: Processed text data.
    """
    corpus = []
    for i in range(0, len(df)):
        # Step 1: Remove non-alphabetic characters
        review = re.sub('[^a-zA-Z]', ' ', df[text_column][i])

        # Step 2: Lowercase the text
        review = review.lower()

        # Step 3: Tokenize the text by splitting into words
        review = review.split()

        # Step 4: Lemmatize and remove stopwords
        review = [lemmatizer.lemmatize(word) for word in review if word not in stop_words]

        # Step 5: Join words back into a single string
        review = ' '.join(review)

        # Append the processed review to the corpus
        corpus.append(review)

    return pd.Series(corpus)

# Example usage
df['processed_message'] = preprocess_text_data(df, 'message')
```

# Text Preprocessing II

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['label'] = le.fit_transform(df['label'])
df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['processed_message'],df['label'],test_size=0.20)
```

### Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=2500,ngram_range=(1,2))

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## TFIDF Vectorizer

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['processed_message'],df['label'],test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=2500,ngram_range=(1,2))

X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
tfidf.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
tfidf_model = MultinomialNB()
```

```python
tfidf_model.fit(X_train,y_train)
```

```python
y_pred=tfidf_model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
